```python
RUN_DELETE_ALL = False  # Set to True to delete all documents in the collections

if RUN_DELETE_ALL:

    import sys
    import os

    sys.path.insert(0, os.path.abspath(".."))

    from configs.config import Config
    from src.mongo_utils import get_collection

    # ==================================================
    # COLLECTIONS
    # ==================================================

    collection_raw = get_collection(
        uri=Config.mongo.URI,
        db_name=Config.mongo.DB_NAME,
        collection_name=Config.mongo.COLLECTION_RAW,
    )

    collection_jobs = get_collection(
        uri=Config.mongo.URI,
        db_name=Config.mongo.DB_NAME,
        collection_name=Config.mongo.COLLECTION_JOBS,
    )

    # ==================================================
    # HITUNG DATA SEBELUM HAPUS
    # ==================================================

    raw_count = collection_raw.count_documents({})
    jobs_count = collection_jobs.count_documents({})

    print(f"Komentar ditemukan : {raw_count:,}")
    print(f"Job ditemukan      : {jobs_count:,}")

    # ==================================================
    # DELETE ALL DOCUMENTS
    # ==================================================

    raw_result = collection_raw.delete_many({})
    jobs_result = collection_jobs.delete_many({})

    print("\n[OK] DATABASE BERHASIL DIBERSIHKAN")

    print(
        f"Komentar terhapus : "
        f"{raw_result.deleted_count:,}"
    )

    print(
        f"Job terhapus      : "
        f"{jobs_result.deleted_count:,}"
    )

else:
    print("DELETE MODE NONAKTIF")
```
